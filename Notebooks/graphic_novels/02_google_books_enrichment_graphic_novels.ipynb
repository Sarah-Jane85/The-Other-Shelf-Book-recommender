# 02 Description Enrichment — Graphic Novels

Goal:
Add book descriptions from Open Library work pages.

Input:
`../Data/Raw/graphic_novels/openlibrary_graphic_novels_raw.csv`

Output:
`../Data/Clean/graphic_novels_with_descriptions.csv`

```python
import requests
import pandas as pd
import time
import json
from pathlib import Path
```

```python
raw_path = Path("../Data/Raw/graphic_novels/openlibrary_graphic_novels_raw.csv")

df = pd.read_csv(raw_path)

df.shape
```

```python
# Display the first few rows of the DataFrame to check the data
df[["title", "author", "ol_key"]].head()
```

```python
# Function to fetch description from Open Library API using the ol_key
def fetch_openlibrary_description(ol_key):
    if pd.isna(ol_key):
        return None

    try:
        url = f"https://openlibrary.org{ol_key}.json"
        response = requests.get(url, timeout=10)

        if response.status_code != 200:
            return None

        data = response.json()
        description = data.get("description")

        if isinstance(description, dict):
            return description.get("value")
        elif isinstance(description, str):
            return description
        else:
            return None

    except Exception:
        return None
```

```python
# Fetch descriptions for all books and store in a new column
descriptions = []

for i, key in enumerate(df["ol_key"]):
    desc = fetch_openlibrary_description(key)
    descriptions.append(desc)

    if i % 100 == 0:
        print(f"Processed {i} / {len(df)}")

    time.sleep(0.2)

df["description"] = descriptions
```

```python
# Check how many descriptions were successfully fetched
df["description"].isna().sum()
```

```python
# Save enriched data to CSV
clean_dir = Path("../Data/Clean")
clean_dir.mkdir(parents=True, exist_ok=True)

output_path = clean_dir / "graphic_novels_with_descriptions.csv"

df.to_csv(output_path, index=False)

print(f"Saved enriched data to: {output_path}")
print(f"Rows saved: {len(df)}")
```
